### Simple Guess Game - Battleship

```python
from random import randint # to activate random numbers function, randint method

board=[] #create a list
for x in range(0,5):# create 5 lists
    board.append(["|O|"]*5)# that contains 5 "|O|" in each of the 5 lists

def print_board(board):
    s = ' ' # initialize a variable to remove punctuation '' and ,
    for i in board:# use the for loop function, print the 5 lists to the bottom
        print ('    ', s.join(i), '\n') # use join method


#create 2 random numbers, each for row and col.
def random_row(board):
    return randint(0, len(board)-1) # return rand numbers between 0 and len(board)-1

def random_col(board):
    return randint(0, len(board)-1)


#assign the rand numbers to the below variables
ship_row = random_row(board)
ship_col = random_col(board)

# printing the intro of the game
print()
print('>>>> WELCOME TO THE GAME <<<<','\n')
#input()
print('        MY BATTLESHIP   ','\n')
print_board(board)
print('*******************************','\n')

input('Press enter to play...')



# start of the game, using for loop in the body statement
for turn in range(3): # range(3) means will iterate 3 times (3 turns)
    print()
    print ("Turn:",turn + 1,'\n')
    # try and except function to catch gamers error
    try:
        guess_row = (int(input('Guess Row: ')) - 1) # -1, so that the user can enter num 1 - 5, instead of 0 - 4
        print()
        guess_col = (int(input('Guess Col: ')) - 1)
    except:
        print()
        print('>>> Oops! Please enter numbers only! <<<','\n')
        break # if the gamers enter non int, the game will end

    # the conditional statements
    if guess_row == ship_row and guess_col == ship_col:
        print()
        print ("CONGRATULATIONS! You sunk my Battleship!",'\n')
        board[guess_row][guess_col] = "|X|"
        print_board(board)
        break
    else:
        if (guess_row < 0 or guess_row > 4) or (guess_col < 0 or  guess_col > 4):
            print ("Oops, that's NOT even in the OCEAN!")
        elif (board[guess_row][guess_col] == "|X|"):#if gamers already make the same guess
            print()
            print ("You GUESSED that one already!",'\n')
            print_board(board)
        else:
            print()
            print (" You MISSED my Battleship!",'\n')
            board[guess_row][guess_col] = "|X|"
            if turn == 2:
                board[ship_row][ship_col] = "|S|"
            print_board(board)
            if turn == 0 or turn < 2:# to print below
                input('Press enter for your next turn...')

print('>>> Come Back & Play Again! <<<', '\n')
print('>>> See You Next Time! <<<')
```
